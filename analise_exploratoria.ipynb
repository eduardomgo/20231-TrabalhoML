### Importando dados

```python
def generate_full_bus_df():
    months = ['Jan', 'Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']

    bus_data = []

    for month in months:
        bus_df_adress = 'database/' + month + '2019.csv'
        df = pd.read_csv(bus_df_adress, sep=";", low_memory=False)
        bus_data.append(df)

    df_bus_data = pd.concat(bus_data, ignore_index=True)

    return df_bus_data
```

```python
df_bus = generate_full_bus_df()
```

## Trabalhando com datetime

```python
# df_bus

df_bus['DataIni'] = pd.to_datetime(df_bus['DataIni'], format='%d/%m/%Y %H:%M:%S')
df_bus['DataFim'] = pd.to_datetime(df_bus['DataFim'], format='%d/%m/%Y %H:%M:%S')
df_bus['HoraIni'] = pd.to_datetime(df_bus['HoraIni'], format='%d/%m/%Y %H:%M:%S')
df_bus['HoraFim'] = pd.to_datetime(df_bus['HoraFim'], format='%d/%m/%Y %H:%M:%S')
df_bus['DuraçãoViagem'] = pd.to_datetime(df_bus['DuraçãoViagem'], format='%d/%m/%Y %H:%M:%S')

```

```python
df_bus
```

```python
df_bus.describe(datetime_is_numeric=True)#.apply(lambda s: s.apply('{0:.2f}'.format))
```

Obs. HoraIni e HoraFim não têm data relevante associada, apenas horário. O mesmo vale para DuraçãoViagem. Desta forma, esses três campos podem ser substituidos apenas pelas horas.

```python
df_bus['HoraIni'] = df_bus['HoraIni'].dt.time
df_bus['HoraFim'] = df_bus['HoraFim'].dt.time
df_bus['DuraçãoViagem'] = df_bus['DuraçãoViagem'].dt.time
```

```python
df_bus
```
